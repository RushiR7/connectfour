```python
class ConnectFour:
    def __init__(self,row,col,connect):
        self.row=row
        self.col=col
        self.connect=connect
        self.Total_pieces=row*col
        self.current_pieces=0
        self.pannel=[[None for i in range(self.col)] for j in range(self.row)]

    def show_pannel(self):
        for i in range(self.row):
            for j in range(self.col):
                if j==0:
                    print("| ",end="")
                if self.pannel[i][j]==None:
                    print(" ",end=" | ")
                else:
                    print(self.pannel[i][j],end=" | ")
            print()
            for k in range(self.row):

                print("-----",end="")
            print()

    def add_piece(self,col,code):

        for i in range(self.row,-1,-1):

            if self.pannel[i-1][col-1] == None:
                self.pannel[i-1][col-1]=code
                self.current_pieces +=1
                p=self.current_pieces
                return i-1,col-1,p

    def check_win(self,x,y,code):
        #check veritical downword 
        if self.check_vertically_down(x,y,code)==True:
            return True
        elif self.check_Horizontally(x,y,code)==True:
            return True
        elif self.check_incline_forword(x,y,code)==True:
            return True
        elif self.check_incline_backword(x,y,code)==True:
            return True


        return False 

        
        
    def check_vertically_down(self,x,y,code):
        x=x+1
        c=1
        while x<self.row and self.pannel[x][y]==code :
            c=c+1
            x=x+1
        
        if c==self.connect:
            return True
        return False

    def check_Horizontally(self,x,y,code):
        c=1
        #Counting right side of current piece
        y1=y+1
        while y1<self.col and self.pannel[x][y1]==code :
            c=c+1
            y1=y1+1
        
        #Counting left side of current piece
        y2=y-1
        while y2>=0 and self.pannel[x][y2]==code :
            c=c+1
            y2=y2-1

        if c>=self.connect:
            return True
        return False
    
    def check_incline_forword(self,x,y,code):

        c=1
        #checking 45 deg upword direction
        x1=x-1
        y1=y+1
        while x1>=0 and y1<self.col and self.pannel[x1][y1]==code:
            x1=x1-1
            y1=y1+1
            c=c+1
        
        #checking 45 deg downword direction

        x2=x+1
        y2=y-1
        while x2<self.row and y2>=0 and self.pannel[x2][y2]==code:
            x2=x2+1
            y2=y2-1
            c=c+1

        if c>=self.connect:
            return True
        return False

    def check_incline_backword(self,x,y,code):

        c=1
        #checking 45 deg upword direction
        x1=x-1
        y1=y-1
        while x1>=0 and y1>=0 and self.pannel[x1][y1]==code:
            x1=x1-1
            y1=y1-1
            c=c+1
        
        #checking 45 deg downword direction

        x2=x+1
        y2=y+1
        while x2<self.row and y2<self.col and self.pannel[x2][y2]==code:
            x2=x2+1
            y2=y2+1
            c=c+1

        if c>=self.connect:
            return True
        return False
```
